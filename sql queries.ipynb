# 1.Order Status
Order status is a crucial metric for evaluating the efficiency of order processing and overall operational performance of an e-commerce company. In this analysis, we will examine the distribution of order statuses for Q1-Q2 2023. The order statuses typically include categories such as "Shipped," "Completed," "Cancelled," and "Returned."

By studying the distribution of order statuses, we can gain insights into the effectiveness and efficiency of the order fulfillment process. We will be able to identify any potential issues or bottlenecks that may have occurred during this period.

The queries that i use are :

 SELECT o.status, COUNT(*) AS order_count

 FROM `bigquery-public-data.thelook_ecommerce.orders` o

 LEFT JOIN `bigquery-public-data.thelook_ecommerce.order_items` oi ON o.order_id = oi.order_id

 LEFT JOIN `bigquery-public-data.thelook_ecommerce.products` p ON oi.product_id = p.id

 WHERE o.created_at >= TIMESTAMP("2023-01-01") AND o.created_at <= TIMESTAMP("2023-06-30")
 
 GROUP BY o.status


 ORDER BY order_count DESC;

# 3.Distribution Center Performance
Distribution centers play a crucial role in an e-commerce company's operations, as they are responsible for order fulfillment and shipping. In this analysis, we will assess the performance of each distribution center in terms of order fulfillment and shipping. By evaluating their performance, we can identify any disparities and determine areas that require improvement.

the queries that i use are :

SELECT ii.product_distribution_center_id, COUNT(*) AS product_count

FROM `bigquery-public-data.thelook_ecommerce.inventory_items` ii

JOIN `bigquery-public-data.thelook_ecommerce.order_items` oi ON ii.product_id = oi.product_id

JOIN `bigquery-public-data.thelook_ecommerce.orders` o ON oi.order_id = o.order_id

WHERE o.status IN ('Shipped', 'Completed')

  AND EXTRACT(YEAR FROM o.created_at) = 2023

  AND EXTRACT(QUARTER FROM o.created_at) <= 2

GROUP BY ii.product_distribution_center_id

ORDER BY product_count DESC;

# 4.Top Destination Countries
Understanding the top destination countries to which The Look shipped its products during Q1-Q2 2023 is essential for assessing the company's international reach and identifying potential growth opportunities. By analyzing the shipment data and identifying the countries with the highest order volumes, this analysis will provide valuable insights into the company's global performance and highlight regions where The Look has established a strong presence.

the queries that i use are :



WITH
  cust AS (

    SELECT 

      DISTINCT oi.user_id,

      u.country AS country

    FROM `bigquery-public-data.thelook_ecommerce.order_items` AS oi

    INNER JOIN `bigquery-public-data.thelook_ecommerce.users` AS u  

    ON oi.user_id = u.id

    WHERE oi.status NOT IN ('Cancelled','Returned')

      AND EXTRACT(YEAR FROM oi.created_at) = 2023

      AND EXTRACT(QUARTER FROM oi.created_at) <= 2

    GROUP BY 1, 2

  )
  

SELECT

  c.country,


  COUNT(DISTINCT c.user_id) AS customers_count

FROM cust AS c

GROUP BY 1

ORDER BY 2 DESC

LIMIT 5;

# 5.Cancelation Status based on Gender:
Understanding the cancelation status of orders based on gender can provide valuable insights into customer behavior and potential factors that influence order cancelations. By analyzing the data and examining if there is any correlation between gender and order cancelation, we can uncover patterns and trends that may help The Look make informed decisions to optimize their operations and customer experience.

the queries that i sue are :

SELECT gender, COUNT(*) AS canceled_count

FROM `bigquery-public-data.thelook_ecommerce.orders`

WHERE status = 'Cancelled'

GROUP BY gender;



During our analysis of order cancelation based on gender, we have observed that there is a difference in the cancelation percentages between male and female customers. This finding suggests that gender may play a role in influencing the decision to cancel an order.

It is interesting to note that both male and female customers exhibit a similar pattern in terms of cancelation rates. This implies that factors beyond gender, such as product dissatisfaction, delivery issues, or changes in personal circumstances, may have a more significant impact on the cancelation decision.
